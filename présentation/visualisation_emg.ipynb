```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the EMG and flexion angle data from the .txt file
file_path = "/Users/Julie/Desktop/projet kyushu/SEMG_DB1/N_TXT/1Nmar.txt"
col_names = ['RF', 'BF', 'VM', 'ST', 'FX']  # Muscle names + flexion angle
dataset = pd.read_csv(file_path, sep="\t", header=None, names=col_names, engine='python')

# Convert data to numeric 
dataset = dataset.apply(pd.to_numeric, errors='coerce')
dataset.dropna(inplace=True)  # Remove rows with NaNs to avoid plotting issues

# Create a time axis 
time = list(range(len(dataset)))

# Set custom colors for each signal 
colors = {
    'RF': 'blue',      # Rectus Femoris
    'BF': 'green',     # Biceps Femoris
    'VM': 'red',       # Vastus Medialis
    'ST': 'magenta',   # Semitendinosus
    'FX': 'cyan'       # Knee flexion angle
}

# Create subplots 
fig, axs = plt.subplots(5, 1, figsize=(12, 18))  

# Plot each signal on its own subplot
for i, label in enumerate(col_names):
    axs[i].plot(time, dataset[label], color=colors[label], label=f'{label} vs Time')
    axs[i].set_title(f'Dataset 1 - Graphique de {label} en fonction du temps') 
    axs[i].legend(loc='upper right')
    axs[i].set_xlabel('Time')
    axs[i].set_ylabel(label)
    axs[i].grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()

# Show final visualization
plt.show()

```
